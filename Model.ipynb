```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Understanding the Data

```python
df = pd.read_csv("data.csv")
```

```python
df
```

# Drop Unnecessary columns

```python
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
# Encode target variable
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
df.info()
```

# Exploratory Data analysis (EDA)

```python
# Set plot styles
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
```

```python
# 1. Count Plot for Diagnosis
sns.countplot(x='diagnosis', data=df, palette='Set2')
plt.title('Distribution of Diagnosis (0: Benign, 1: Malignant)')
plt.show()
```

In the given data, almost **60%** of the records are benign(0).\
**40%** of the records are malignant(1).

```python
# 2. Pairplot of selected features
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'diagnosis']
sns.pairplot(df[selected_features], hue='diagnosis', palette='Set1')
plt.show()
```

**1.** The cancer is **Benign** as long as the radius is **Smaller**.\
**2.** As the radius increases, The chances of **Malignancy** **Increases**.\
**3.** **Texture** has not a strong relationship with the diagonosis.\
**4.** If the **Perimeter mean** is small, Most of the chances are of benign. As the **Perimeter mean** increases, the chances of Malignancy **Increases**.\
**5.** If the **Area** is small, Most of the chances are of benign. As the **Area** increases, the chances of Malignancy **Increases**.\
**6.** **Smoothness** has not a found to have a strong relationship with the diagonosis.\

**All** the selected features almost forms linear Decision Boundaries with each other.

```python
# 3. Correlation Heatmap (top 10 most correlated with diagnosis)
correlations = df.corr()['diagnosis'].abs().sort_values(ascending=False)[1:11]
plt.figure(figsize=(10, 8))
sns.heatmap(df[correlations.index].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Top 10 Correlated Features with Diagnosis')
plt.show()
```

1. **Concave points worst** has a very high correlation with top 10 with an average of almost 0.85.
2. **Perimeter worst** also has a very high correlation with top 10 with the lowest with concacity worst of about 0.62.
3. **Concave points worst** has a very high correlation with top 10 with an average of almost 0.83.
4. **Radius Worst** also has a very high correlation with top 8 and a little lower with **Concavity mean** and **Concavity worst**.
